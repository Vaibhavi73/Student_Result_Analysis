```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("Desktop/Data Analyst ExcelR/Project/Student Exam Score Python/archive (2)/Expanded_data_with_more_features.csv")
print(df.head())
```

# Drop Unnamed Column 

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
df.head()
```

# Gender Distribution

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data = df, x = "Gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show
```

```python
# From the above chart we have analyzed that:- the number of females in the data is more than the number of men 
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parents Education & Students Score")
plt.show()
```

```python
## From the above chart we have concluded that the education of the parents have a good impact on their score 
```

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parents Marital Status & Students Score")
plt.show()
```

```python
## From the above chart we have concluded that there is no/negligible impact on the student score due to there parents marital status  
```

# Distribution of Ethnic Groups 

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
print(groupA)
```

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mlist)

plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution Of Ethnic Groups")
plt.show()
```
